```python
# 台形の各辺の長さ（cm）
upper_base = 10   # 上辺
lower_base = 20   # 下辺
height = 5        # 高さ

# 台形の面積を計算
area = (upper_base + lower_base) * height / 2

# 結果を表示
print(f"台形の面積は {area} cm² です。")
```
